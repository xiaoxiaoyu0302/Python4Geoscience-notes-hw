### 1. Plot sin wave harmonics (Tuesday's in class project)

Plot `sin(n*x)` over the range 0 to $\pi$ for values of $n=1,...,5$. Label $x$ values at $0$, $\pi/2$, and $\pi$; label $y$ values at -1, 0, and 1. Create a legend that labels each line $mode=n$ for each value of $n$ plotted.

Label the y-axis sin($nx$), and the x-axis '$x$'. Use math fonts for $n$ and $x$, but not for 'sin', which is traditionally rendered in a Roman (upright) font. 

```python
# YOUR CODE HERE
x=np.linspace(0,np.pi)
n=[1,2,3,4,5]
colors = ('b', 'g', 'r', 'c', 'm', 'y', 'k')
linestyles = ['-', '-', '--', ':', '-.','dotted','dotted']
fig = plt.figure(figsize=(10, 4))
ax1 = fig.add_subplot(1, 1, 1)

for ni, ls, color in zip(n, linestyles, colors):
    label = 'model = {}'.format(ni)
    ax1.plot(x,np.sin(ni*x), ls=ls, color = color, label= label)

ax1.set_title('y=sin(nx)')
ax1.legend(loc='lower left')
# xticklocs = [0,0.5*np.pi, np.pi]
# xticklabels = ["0", ""]
    

```
